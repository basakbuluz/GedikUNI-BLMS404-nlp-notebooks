# TF-IDFâ€™nin Python ile UygulanmasÄ±

**AdÄ±m 1: Gerekli KÃ¼tÃ¼phaneleri Ä°Ã§e Aktarma**

Bu adÄ±mda, metin verisini TF-IDF vektÃ¶rlerine dÃ¶nÃ¼ÅŸtÃ¼rmek iÃ§in gerekli olan scikit-learn kÃ¼tÃ¼phanesini iÃ§e aktaracaÄŸÄ±z.

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

**AdÄ±m 2: DokÃ¼manlardan Metinleri Toplama ve Corpus OluÅŸturma**

Bu adÄ±mda, analiz etmek istediÄŸimiz metinleri bir araya getirerek bir corpus (metin koleksiyonu) oluÅŸturacaÄŸÄ±z.

Corpus, NLP iÅŸlemlerinin baÅŸlangÄ±Ã§ noktasÄ±dÄ±r. TF-IDF hesaplayabilmek iÃ§in, metinlerin bir liste (list) formatÄ±nda tanÄ±mlanmasÄ± gerekir.

````python
d0 = 'The cat sat on the mat'
d1 = 'The dog played in the park'
d2 = 'Cats and dogs are great pets'
string = [d0, d1, d2]
````

**AdÄ±m 3: TF-IDF DeÄŸerlerini Hesaplama**

Bu adÄ±mda, corpus iÃ§inde bulunan metinleri TF-IDF vektÃ¶rlerine dÃ¶nÃ¼ÅŸtÃ¼receÄŸiz. Bunun iÃ§in scikit-learn kÃ¼tÃ¼phanesindeki TfidfVectorizer sÄ±nÄ±fÄ±nÄ± kullanacaÄŸÄ±z.

TfidfVectorizer, metin Ã¼zerinde otomatik olarak ÅŸu iÅŸlemleri gerÃ§ekleÅŸtirir:

- Metni kelimelere ayÄ±rÄ±r (tokenization)
- Vocabulary oluÅŸturur
- Term Frequency (TF) hesaplar
- Inverse Document Frequency (IDF) hesaplar
- ve her kelime iÃ§in TF-IDF skorunu Ã¼retir


````python
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(string)
````

**AdÄ±m 4: IDF DeÄŸerlerini GÃ¶rÃ¼ntÃ¼leme**

````python
print('\nidf values:')
for ele1, ele2 in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(ele1, ':', ele2)
````

**AdÄ±m 5: TF-IDF DeÄŸerlerini ve Ä°ndekslerini GÃ¶rÃ¼ntÃ¼leme**

````python
print('\nWord indexes:')
print(tfidf.vocabulary_)
print('\ntf-idf value:')
print(result)
print('\ntf-idf values in matrix form:')
print(result.toarray())
````
